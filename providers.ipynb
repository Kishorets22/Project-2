3.How many food providers and receivers are there in each city?


```python
cursor.execute("""SELECT 
    City,
    COUNT(id) AS total_providers,
    COUNT(Receiver_ID) AS total_receivers
FROM (
    SELECT City, id, NULL AS Receiver_ID FROM providers
    UNION ALL
    SELECT City, NULL AS id, Receiver_ID FROM recievers
) AS combined
GROUP BY City; """)
result=cursor.fetchall()
pd.DataFrame(result,columns=['city','total_providers','total_receivers'])
```

4.Which receivers have claimed the most food?

```python
cursor.execute(""" SELECT 
    r.Receiver_ID,
    r.Name AS Receiver_Name,
    r.Type AS Receiver_Type,
    COUNT(c.Claim_ID) AS Total_Claims
FROM 
    claims c
JOIN 
    recievers r ON c.Receiver_ID = r.Receiver_ID
GROUP BY 
    r.Receiver_ID, r.Name, r.Type
ORDER BY 
    Total_Claims DESC;

""")
result=cursor.fetchall()
pd.DataFrame(result,columns=['Receiver_ID','Receiver_name','Receiver_type','Total_claims'])
```

 10. What percentage of food claims are completed vs. pending vs. canceled?

```python
cursor.execute("""SELECT 
    Status,
    COUNT(*) AS Count,
    ROUND((COUNT(*) / (SELECT COUNT(*) FROM claims) * 100), 2) AS Percentage
FROM 
    claims
GROUP BY 
    Status;""")
result=cursor.fetchall()
pd.DataFrame(result,columns=["Status","count","Percentage"])
```

12.What is the total quantity of food donated by each provider?

13. What is the average quantity of food claimed per receiver?


```python
cursor.execute(""" select c.Receiver_ID, avg(f.Quantity) as Avg 
               from claims c 
               join 
               foodlisting f 
               group by c.Receiver_ID 
               order by Avg desc
""")
result=cursor.fetchall()
pd.DataFrame(result,columns=['Receiver_ID','Average_Quantity'])
```

14. Providers with highest average quantity per listing

```python
# Average quantity per listing by provider (SQL)
cursor.execute("""SELECT Provider_ID, AVG(Quantity) AS Avg_Quantity
               FROM foodlisting
               GROUP BY Provider_ID
               ORDER BY Avg_Quantity DESC
""")
result = cursor.fetchall()
pd.DataFrame(result, columns=['Provider_ID','Avg_Quantity'])
```

15. Providers with largest variety of food items (distinct Food_Name)

```python
# Providers with largest variety of food items (SQL)
cursor.execute("""SELECT Provider_ID, COUNT(DISTINCT Food_Name) AS Distinct_Food_Count
               FROM foodlisting
               GROUP BY Provider_ID
               ORDER BY Distinct_Food_Count DESC
""")
result = cursor.fetchall()
pd.DataFrame(result, columns=['Provider_ID','Distinct_Food_Count'])
```

17. Distribution of provider types across cities

18. Claims status breakdown overall and by provider type

```python
# Claims status breakdown overall and by provider type (SQL)
cursor.execute("""SELECT Status, COUNT(*) AS Count, ROUND((COUNT(*)/(SELECT COUNT(*) FROM claims)*100),2) AS Percentage
               FROM claims
               GROUP BY Status
""")
result = cursor.fetchall()
status_counts = pd.DataFrame(result, columns=['Status','Count','Percentage'])
# Breakdown by provider type
cursor.execute("""SELECT f.Provider_Type, c.Status, COUNT(*) AS Count
               FROM claims c
               JOIN foodlisting f ON c.Food_ID = f.Food_ID
               GROUP BY f.Provider_Type, c.Status
               ORDER BY f.Provider_Type, Count DESC
""")
result = cursor.fetchall()
by_provider_type = pd.DataFrame(result, columns=['Provider_Type','Status','Count'])
status_counts
by_provider_type
```

19. Cancellation rate per provider (cancelled / total claims for provider)

```python
# Cancellation rate per Provider_ID
claims_food = data_claim.merge(data_food, on='Food_ID', how='left')
total_by_provider = claims_food.groupby('Provider_ID').size().reset_index(name='Total_Claims')
cancelled_by_provider = claims_food[claims_food['Status'].str.lower()=='cancelled'].groupby('Provider_ID').size().reset_index(name='Cancelled')
cancel_rate = total_by_provider.merge(cancelled_by_provider, on='Provider_ID', how='left').fillna(0)
cancel_rate['Cancelled'] = cancel_rate['Cancelled'].astype(int)
cancel_rate['Cancellation_Rate'] = (cancel_rate['Cancelled'] / cancel_rate['Total_Claims'] * 100).round(2)
cancel_rate.sort_values('Cancellation_Rate', ascending=False)
```

20. Claims per day (time-series)

```python
# Parse Timestamp and aggregate claims per day
data_claim['Timestamp'] = pd.to_datetime(data_claim['Timestamp'], errors='coerce')
claims_by_day = data_claim.groupby(data_claim['Timestamp'].dt.date).size().reset_index(name='Count').rename(columns={'Timestamp':'Date'})
claims_by_day = claims_by_day.sort_values(by='Timestamp') if 'Timestamp' in claims_by_day.columns else claims_by_day
claims_by_day
```
